# Notebook 02 - Pré-processamento
Limpeza e normalização dos tweets coletados.

```
import pandas as pd
import re

# Exemplo: carregar tweets brutos
hb20 = pd.read_csv('../data/raw/HB20_tweets.csv')
hb20.head()
```

```
def limpar_texto(texto):
    texto = re.sub(r'http\S+', '', texto)
    texto = re.sub(r'[^a-zA-ZÀ-ÿ ]', '', texto)
    return texto.lower().strip()
```

```
hb20['clean'] = hb20['content'].apply(limpar_texto)
hb20.to_csv('../data/processed/HB20_tweets_clean.csv', index=False)
hb20.head()
```
