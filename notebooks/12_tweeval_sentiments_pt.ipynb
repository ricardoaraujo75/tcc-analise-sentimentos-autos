```python
# 1) Imports
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch
```

```python
# 2) Modelo público em PT-BR (pysentimiento)
MODEL_PT = "pysentimiento/bertweet-pt-sentiment"
```

```python
# 3) Carregar tokenizer/model com fallback
try:
    # instalar sentencepiece se necessário, e forçar tokenizer lento ao detectar problemas
    tokenizer = AutoTokenizer.from_pretrained(MODEL_PT, use_fast=True)
    model = AutoModelForSequenceClassification.from_pretrained(MODEL_PT)
except Exception as e:
    print("Falha ao carregar tokenizer/model (tentando fallback com use_fast=False). Erro:", e)
    tokenizer = AutoTokenizer.from_pretrained(MODEL_PT, use_fast=False)
    model = AutoModelForSequenceClassification.from_pretrained(MODEL_PT)
```

```python
# 4) Pipeline
sentiment = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, device=0 if torch.cuda.is_available() else -1)
```

```python
# 5) Teste rápido
frases = [
    "O HB20 é econômico, mas o acabamento poderia ser melhor.",
    "O Onix é muito confortável e tem ótimo custo-benefício.",
    "Não gostei do desempenho do Kwid na estrada."
]

for f in frases:
    print(f, "->", sentiment(f))
```
