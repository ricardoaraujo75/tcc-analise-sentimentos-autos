# Notebook 05 - Dashboard Demo
Pré-visualização do dashboard em Jupyter antes do Streamlit.

```
import pandas as pd
import matplotlib.pyplot as plt

hb20 = pd.read_csv('../data/processed/HB20_tweets_clean.csv')
hb20['sentiment'] = ['POS','NEG','NEU'] * (len(hb20)//3)

hb20['sentiment'].value_counts().plot(kind='bar')
plt.title('Distribuição de sentimentos HB20')
plt.show()
```
